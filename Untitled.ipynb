```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# تعریف محیط سه‌بعدی
grid_size = 10
environment = np.zeros((grid_size, grid_size, grid_size))  # محیط خالی سه‌بعدی

# قرار دادن منابع انرژی به صورت تصادفی
num_resources = 5
resource_positions = np.random.randint(0, grid_size, (num_resources, 3))
for pos in resource_positions:
    environment[pos[0], pos[1], pos[2]] = 1  # 1 نشان‌دهنده منبع انرژی است
    
class Robot:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        self.energy = 100
        self.path = [(x, y, z)]  # مسیر حرکت ربات

    def move(self, new_x, new_y, new_z):
        if 0 <= new_x < grid_size and 0 <= new_y < grid_size and 0 <= new_z < grid_size:
            self.x = new_x
            self.y = new_y
            self.z = new_z
            self.energy -= 1  # کاهش انرژی با هر حرکت
            self.path.append((new_x, new_y, new_z))  # اضافه کردن موقعیت جدید به مسیر

    def sense_resources(self, environment):
        resources = []
        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                for dz in [-1, 0, 1]:
                    if 0 <= self.x + dx < grid_size and 0 <= self.y + dy < grid_size and 0 <= self.z + dz < grid_size:
                        if environment[self.x + dx, self.y + dy, self.z + dz] == 1:
                            resources.append((self.x + dx, self.y + dy, self.z + dz))
        return resources
# ایجاد ربات‌ها
robots = [Robot(np.random.randint(0, grid_size), np.random.randint(0, grid_size), np.random.randint(0, grid_size)) for _ in range(2)]

# شبیه‌سازی
time_steps = 100
for t in range(time_steps):
    print(f"Time step {t}")
    for robot in robots:
        if robot.energy <= 0:
            continue  # ربات انرژی ندارد و نمی‌تواند حرکت کند

        # تشخیص منابع در نزدیکی
        resources = robot.sense_resources(environment)
        if resources:
            target = resources[0]  # به سمت اولین منبع حرکت کند
            dx = target[0] - robot.x
            dy = target[1] - robot.y
            dz = target[2] - robot.z

            # حرکت به سمت منبع
            if dx != 0:
                robot.move(robot.x + np.sign(dx), robot.y, robot.z)
            elif dy != 0:
                robot.move(robot.x, robot.y + np.sign(dy), robot.z)
            elif dz != 0:
                robot.move(robot.x, robot.y, robot.z + np.sign(dz))

            # اگر به منبع رسید، انرژی افزایش یابد
            if (robot.x, robot.y, robot.z) == target:
                robot.energy += 20
                environment[target[0], target[1], target[2]] = 0  # منبع مصرف شده است

        print(f"Robot at ({robot.x}, {robot.y}, {robot.z}) with energy {robot.energy}")

    # بررسی پایان شبیه‌سازی
    if all(robot.energy <= 0 for robot in robots):
        print("All robots are out of energy!")
        break

# نمایش گراف سه‌بعدی
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("3D Robot Movement Visualization")

# نمایش منابع انرژی
for pos in resource_positions:
    ax.scatter(pos[0], pos[1], pos[2], color='green', s=100, label='Resource' if pos[0] == resource_positions[0][0] else "")

# نمایش مسیر حرکت ربات‌ها
colors = ['red', 'blue']  # رنگ‌های مختلف برای ربات‌ها
for i, robot in enumerate(robots):
    path = np.array(robot.path)
    ax.plot(path[:, 0], path[:, 1], path[:, 2], marker='o', color=colors[i], label=f'Robot {i+1}')
    ax.scatter(robot.path[-1][0], robot.path[-1][1], robot.path[-1][2], marker='X', color=colors[i], s=100)  # موقعیت نهایی

# تنظیمات گراف
ax.set_xlabel("X Coordinate")
ax.set_ylabel("Y Coordinate")
ax.set_zlabel("Z Coordinate")
ax.set_xticks(np.arange(0, grid_size, 1))
ax.set_yticks(np.arange(0, grid_size, 1))
ax.set_zticks(np.arange(0, grid_size, 1))
ax.grid(True)
ax.legend()

plt.show()
```
